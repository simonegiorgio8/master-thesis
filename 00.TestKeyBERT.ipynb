Max Sum Distance: per diversificare il risultato:
"we take the 2 x top_n most similar words/phrases to the document. Then, we take all top_n combinations from the 2 x top_n words and extract the combination that are the least similar to each other by cosine similarity.

```python
model.extract_keywords(doc[:512],keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_maxsum=True, nr_candidates=20, top_n=5)
```

Maximal Marginal Relevance
To diversify the results, we can use Maximal Margin Relevance (MMR) to create keywords / keyphrases which is also based on cosine similarity.

```python
model.extract_keywords(doc, keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_mmr=True, diversity=0.7)
```

```python
parag_len = 512
segmenti = [doc[i:i+parag_len] for i in range(0, len(doc), parag_len)]

# Estraggo keywords da ogni segmento
keywords_tot = []
for segmento in segmenti:
    kws = model.extract_keywords(segmento, keyphrase_ngram_range=(1, 3), #stop_words='english',
                              #use_maxsum=True, top_n=3,
                              use_mmr=True, diversity=0.7)
    keywords_tot.extend(kws)

print(keywords_tot)
```

```python
ordered_kws_tot = sorted(keywords_tot, key=lambda x: x[1], reverse=True)
top_kws = ordered_kws_tot[:10]
```

```python
nltk.download('stopwords')
stopwords_set = set(stopwords.words('english'))
```

```python
def is_trigram_present(sentence, tri_gram):
    # genero la frase in analisi senza stopwords
    sentence_words = [parola.lower() for parola in nltk.word_tokenize(sentence) if parola.lower() not in stopwords_set]

    trigram_words = [parola.lower() for parola in tri_gram.split()]

    # Cerca la sequenza delle parole del trigram nella frase
    indice_trigram = 0
    for parola in sentence_words:
        if parola == trigram_words[indice_trigram]:
            indice_trigram += 1
            if indice_trigram == len(trigram_words):
                return True
        elif indice_trigram>0:
          indice_trigram=0 #devo catturare il fatto che la next parola della sentence è diversa da quella corrente del trigram
    return False



tri_grammi_e_frase = {tri_gram: [] for tri_gram in top_search}  # dizionario solo per vedere cosa estrae
includedIndices = set()

# Per ogni trigram trovato
for tri_gram in top_search:
    # Per ogni frase
    for indice, sentence in enumerate(docSentences):
        # Verifica la presenza del trigram nella frase, se indice della frase è già incluso non la re-inserisco.
        if indice not in includedIndices and is_trigram_present(sentence, tri_gram):
            tri_grammi_e_frase[tri_gram].append([sentence, indice]) # indice lo uso successivamente per riordinare frasi
            includedIndices.add(indice)

for tri_gram, frasi in tri_grammi_e_frase.items():
    print(f"Tri-gram: {tri_gram}")
    print(f"Frasi: {frasi}")
    print()
```

```python
sentence_indice_list = [(sentence, indice) for trigram, sentences_and_indices in tri_grammi_e_frase.items() for sentence, indice in sentences_and_indices]
```
